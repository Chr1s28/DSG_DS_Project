# Beurteilung der Zielerreichung

## Data Mining- / Modellierungs-Ziele

Die in der Projektcharta definierten quantitativen Erfolgsmetriken wurden mit dem finalen Modell (XGBoost mit Songtext-Embeddings) gemischt erreicht.

- **Metrik $R^2$ (Erklärungsgehalt):**
    - *Zielwert:* ≥ 0.4 
    - *Erreicht:* 0.242   
    - *Beurteilung:* Ziel verfehlt. Das Modell kann rund 24% der Varianz in der Popularität erklären. Dies ist eine deutliche Steigerung gegenüber der Baseline (5.4%), bleibt jedoch unter dem angestrebten Zielwert von 40%. Dies deutet darauf hin, dass ein erheblicher Teil der Popularität nicht durch intrinsische Song-Eigenschaften (Audio & Text) erklärt werden kann.

- **Metrik RMSE (Prognosegenauigkeit):**
    - *Zielwert:* ≤ 15 (Projektziel) bzw. ≤ 10 (Data Mining Ziel) 
    - *Erreicht:* 11.69
    - *Beurteilung:* Teilweise erreicht. Das übergeordnete Projektziel einer durchschnittlichen Abweichung von maximal 15 Punkten auf der Popularitätsskala wurde erfolgreich unterboten. Das ambitioniertere technische Ziel von ≤ 10 wurde jedoch knapp verfehlt.

## Projektziele

Das primäre Projektziel war es, dem Talent Manager und den Künstlern nützliche Prognosen zu liefern, um das Erfolgspotenzial von Songs einzuschätzen.

**Erfüllung der Anforderungen:**
Das entwickelte Modell liefert mit einem RMSE von 11.69 eine solide Indikation für die zu erwartende Popularität. Es eignet sich gut, um "tote" Tracks (ohne Erfolgsaussichten) frühzeitig zu identifizieren und vom manuellen Review-Prozess auszuschliessen.

**Identifizierte Lücken:**
Die Analyse hat gezeigt, dass rein musikalische und textliche Inhalte den kommerziellen Erfolg nur bedingt vorhersagen können. Wichtige externe Einflussfaktoren, die in der Realität massgeblich für einen Hit verantwortlich sind - wie Marketingbudget, Social-Media-Trends (z.B. TikTok) oder die Bekanntheit des Künstlers - konnten im aktuellen Modell aufgrund fehlender Daten nicht berücksichtigt werden.

**Fazit für den Auftraggeber:** Das Tool bietet eine wertvolle, datenbasierte Zweitmeinung und Filterfunktion, sollte jedoch nicht als alleiniges Entscheidungskriterium für Investitionen genutzt werden, sondern als Ergänzung zur menschlichen Expertise des Talent Managers.

# Empfehlungen Nächste Schritte

Aus der Beurteilung ergeben sich folgende Empfehlungen für das weitere Vorgehen:

## Erweiterung des Domänenwissens

Da intrinsische Song-Eigenschaften allein den Erfolg nur begrenzt erklären, sollte der Fokus verstärkt auf externe Marktdynamiken wie Marketingbudgets und Social-Media-Trends (z.B. TikTok) gelegt werden.


## Adaption der Projektziele

Das Ziel einer exakten Vorhersage sollte zugunsten eines Filterinstruments angepasst werden. Das Modell eignet sich primär zur "Negative Selection", um Songs ohne Potenzial frühzeitig auszusortieren, statt exakte Hit-Scores zu garantieren.


## Datenerweiterung

Integration externer Datenquellen zur Steigerung der Vorhersagekraft:

- **Künstler-Metriken:** Historische Daten wie Follower-Zahlen und vergangene Chart-Erfolge.
- **Kontext-Daten:** Informationen über Playlist-Inklusionen oder Radio-Airplay.
- **Erweiterte Audio-Features:** Nutzung von [AcousticBrainz](https://acousticbrainz.org/), um detaillierte Low-Level-Spektraldaten sowie Stimmungs- und Genre-Informationen (via Essentia Toolkit) zu erhalten, die über die Standard-Spotify-Features hinausgehen.


## Rechtliche und ethische Aspekte

- **Plattform-Risiko:** Die Abhängigkeit von den Spotify-Nutzungsbedingungen muss überwacht werden.
- **Bias:** Der Bias der Zielvariable "Popularity" hin zu Mainstream-Musik und Songs mit verfügbaren Lyrics (Genius) ist bei der Anwendung zu berücksichtigen.

## Bereitstellung

- **Nutzung:** Die Resultate dienen dem Talent Manager als operative Entscheidungsunterstützung (Zweitmeinung) zur Priorisierung von Songs.
- **Form:** Übergabe der Software als dokumentierte Jupyter Notebooks an das IT-Team zur technischen Integration.

# Checkpoint-Entscheidung

- **Entscheidungsträger:** Aristos Panteli (Sponsor, Bandwidth Beats)
- **Ergebnis:** Freigabe für eingeschränkten Pilotbetrieb & Initiierung Phase 2
- **Begründung:**
  Die Entscheidung basiert auf dem gemischten Resultat der Zielerreichung. Da die geforderte Prognosegenauigkeit erfolgreich unterboten wurde (RMSE 11.69 gegenüber Zielwert 15), ist der Prototyp technisch reif für den Einsatz als Filterinstrument ("Negative Selection"). Er bietet dem Talent Manager sofortigen Mehrwert, indem er Songs ohne Erfolgsaussichten zuverlässig identifiziert.

  Da jedoch die für eine "Hit-Garantie" notwendige Erklärungskraft ($R^2$ ≥ 0.4) verfehlt wurde, erfolgt keine Freigabe als alleiniges Entscheidungstool für Signings. Stattdessen wird den Empfehlungen folgend eine Erweiterung des Projektumfangs beschlossen: Die Integration von externen Marktdaten (Social Media) und erweiterten Audio-Features (AcousticBrainz) soll in einer zweiten Phase die Lücke zwischen technischer Analyse und Marktrealität schliessen.
